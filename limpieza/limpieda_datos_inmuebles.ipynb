```python
import pandas as pd
import numpy as np


#importar csv de datos
df = pd.read_csv('../inmuebles24.csv', sep=',', encoding='utf-8')
```

```python
df
```

```python
#limpiar la columna ubicacion eliminando \t
df['ubicacion'] = df['ubicacion'].str.replace('\t', '')
```

```python
ubicacion_split = df['ubicacion'].str.split(', ', expand=True)
df['calle'] = ubicacion_split[0]
df['colonia'] = ubicacion_split[1]
df['alcaldia'] = ubicacion_split[2]

# Filtrar casos con 4 resultados después de la separación
df_filtered = df[df['alcaldia'].notnull()]

# Mostrar el DataFrame resultante
print(df_filtered)
```

```python
df_filtered
```

```python
alcaldias_cdmx = [
    'Álvaro Obregón',
    'Azcapotzalco',
    'Benito Juárez',
    'Coyoacán',
    'Cuajimalpa de Morelos',
    'Cuauhtémoc',
    'Gustavo A. Madero',
    'Iztacalco',
    'Iztapalapa',
    'Magdalena Contreras',
    'Miguel Hidalgo',
    'Milpa Alta',
    'Tláhuac',
    'Tlalpan',
    'Venustiano Carranza',
    'Xochimilco'
]
```

```python
# Filtrar las filas que pertenecen a alguna alcaldía de la CDMX
df_cdmx = df[df['alcaldia'].isin(alcaldias_cdmx)]

# Mostrar el DataFrame resultante
print(df_cdmx)
```

```python
#exporta a csv el dataframe resultante y llamalo inmuebles24_clean.csv
df_cdmx.to_csv('../datasets/inmuebles24_clean.csv', index=False, encoding='utf-8')
```
