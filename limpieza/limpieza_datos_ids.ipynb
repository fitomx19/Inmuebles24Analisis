### Limpieza de datos a archivo shp del indice de desarrollo social.

Importamos librerias

```python
import pandas as pd
import numpy as np


#importar csv de datos
df = pd.read_csv('ids.csv', sep=',')
```

```python
df.head()
```

Solo vamos a utilizar el nombre de la colonia, la delegacion y el indice de desarrollo junto con su coeficiente.

```python
#seleccionar las ultimas 7 columnas de la tabla
df = df.iloc[:, -7:]


```

```python
#renombrar columnas para que sean mas entendibles y no tengan espacios
df.columns = ['CP', 'Municipio', 'Colonia', 'MUNICIPIO', 'COLONIA', 'indice', 'indice_texto']
```

```python
#eliminar columnas que no se usaran como MUNICIPIO y COLONIA
df = df.drop(['Municipio', 'Colonia', 'CP'], axis=1)

```

```python
df.head()
```

```python
#Exportar a csv llamando al archivo "ids_clean.csv"
df.to_csv('ids_clean.csv', index=False, encoding='utf-8')
```
