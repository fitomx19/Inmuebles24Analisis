```python
from selenium.webdriver.common.by import By
from selenium import webdriver
import pandas as pd
import re
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
```

```python
#visualizar el dataset completo sin abreviar
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
```

Parametros y listas

```python
lista_anuncios = []
```

```python
for x in range(1,70):
    driver = webdriver.Safari()
    url = "https://www.inmuebles24.com/departamentos-en-renta-en-ciudad-de-mexico-mas-de-1-bano-mas-de-1-recamara-mas-de-1-estacionamiento-mas-de-35-metros-cuadrados-cubiertos-publicado-hace-menos-de-1-mes-hasta-50-anos-pagina-"+str(x)+".html"
    driver.get(url)
    
    contenedor_anuncios = driver.find_element(By.CLASS_NAME, "postings-container")
    contenedor_anuncios_html = contenedor_anuncios.get_attribute("innerHTML")
    soup = BeautifulSoup(contenedor_anuncios_html, "html.parser")
    anuncios = soup.find_all("div")

    for anuncio in anuncios:
        try:
            link = anuncio.find("a")
            print("❤️ Link ->", link.get("href"))
            href = link.get("href")
            lista_anuncios.append("www.inmuebles24.com" + href)
        except:
            pass
    driver.close()
```

```python
lista = lista_anuncios
```

```python
#eliminar duplicados de la lista
lista = list(dict.fromkeys(lista))
lista
```

```python
lista_html = []
for x in lista:
    driver = webdriver.Safari()
    driver.get("https://" + x)
    
    # Esperar hasta que la página esté completamente cargada (puedes ajustar el tiempo máximo según tus necesidades)
    wait = WebDriverWait(driver, 10)
    wait.until(EC.presence_of_element_located((By.TAG_NAME, 'body')))
    
    lista_html.append(driver.page_source)
    driver.quit()
```

```python
lista_html[0]
```

```python
def scrap():
    # de lista_html extraer el div class price-value , section-location-property
    # y el div class description-text

    lista_precio = []
    lista_ubicacion = []
    lista_tamanio = []
    lista_de_baños = []
    lista_de_estacionamientos = []
    lista_de_recamaras = []
    lista_de_anios = []

    for x in lista_html:
        soup = BeautifulSoup(x, "html.parser")
        precio = soup.find("div", class_="price-value")
        tamanio = soup.find("div", class_="article-container-property")
        ubicacion = soup.find("div", class_="section-location-property")

        # Verificar que todos los elementos necesarios estén presentes
        if all(elem is not None for elem in [precio, tamanio, ubicacion]):
            precio_text = re.sub("[^0-9]", "", precio.text) if precio else None
            tamanio_text = re.sub("[^0-9]", "", tamanio.find("h2").text) if tamanio else None
            ubicacion_text = ubicacion.find("h4").text if ubicacion else None
            # buscar todos los li dentro del ul con id = section-icon-features-property
            # y extraer el texto de cada uno
            lista_caracteristicas = []

            caracteristicas = soup.find("ul", id="section-icon-features-property")

            for x in caracteristicas.find_all("li")[:5]:
                lista_caracteristicas.append(x.text)

            # Asegurémonos de que hay al menos 4 elementos en la lista de características
            if len(lista_caracteristicas) >= 4:
                metros_cuadrados = re.sub("[^0-9]", "", lista_caracteristicas[0])
                num_baños = re.sub("[^0-9]", "", lista_caracteristicas[1])
                num_estacionamientos = re.sub("[^0-9]", "", lista_caracteristicas[2])
                num_recamaras = re.sub("[^0-9]", "", lista_caracteristicas[3])
                antiguedad = re.sub("[^0-9]", "", lista_caracteristicas[4])

                # Ahora puedes utilizar estas variables como necesites
                print("Metros cuadrados:", metros_cuadrados)
                print("Número de baños:", num_baños)
                print("Número de estacionamientos:", num_estacionamientos)
                print("Número de recámaras:", num_recamaras)
                print("Antigüedad:", antiguedad)

            # Verificar que todas las listas tengan información antes de agregar a las listas
            if all(x is not None for x in [precio_text, tamanio_text, ubicacion_text, metros_cuadrados, num_baños, num_estacionamientos, num_recamaras, antiguedad]):
                lista_precio.append(precio_text)
                lista_tamanio.append(tamanio_text)
                lista_ubicacion.append(ubicacion_text)
                lista_de_baños.append(num_baños)
                lista_de_estacionamientos.append(num_estacionamientos)
                lista_de_recamaras.append(num_recamaras)
                lista_de_anios.append(antiguedad)

    diccionario = {"precio": lista_precio,
                   "ubicacion": lista_ubicacion,
                   "tamanio": lista_tamanio,
                   "metros_cuadrados": lista_de_baños,
                   "num_baños": lista_de_estacionamientos,
                   "num_estacionamientos": lista_de_recamaras,
                   "num_recamaras": lista_de_recamaras,
                    "antiguedad": lista_de_anios,
                   }
    df = pd.DataFrame(diccionario)

    return df
```

```python
a = scrap()
```

```python
#crear dataset a partir de otro 
df = pd.DataFrame(a)
```
